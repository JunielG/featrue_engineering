### Detect outliers using percentile

```python
max_threshold = heights['height'].quantile(0.95)
max_threshold
```

```python
heights[heights['height'] > max_threshold]
```

```python
min_threshold = heights['height'].quantile(0.05)
min_threshold
```

```python
heights[heights['height'] < min_threshold]
```

### Remove outliers

```python
heights[(heights['height'] < max_threshold) & (heights['height'] > min_threshold)]
```
